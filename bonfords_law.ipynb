```python
import matplotlib.pyplot as plt #For ploting graph
import scipy
import numpy as np
import math      #For Mathametical calculation
from tabulate import tabulate  #For Table Formulation
import pandas as pd
from scipy import stats

Actual_percentage = []
Benfords_predicted = []

fibo_series = []
leading_dig_value = []
leadind_dig = [1, 2, 3, 4, 5, 6, 7, 8, 9]
count_dig = [0,0,0,0,0,0,0,0,0]
variation_per_dig=[]

Total=0

# this for loop generates an fibonacci series
f = 0
s = 1
for i in range(0, 24):
    fibo_series.append(s)
    k = f + s
    f = s
    s = k


# from the fibonacci series bellow for loop fetches the leading digit value
for i in range(0, len(fibo_series)):
    fibo_number = fibo_series[i]
    digit=0
    while fibo_number != 0:
        digit = fibo_number % 10
        fibo_number=fibo_number//10
    leading_dig_value.append(digit)

# Next for loop counts the Leading digit from leading_dig_value[] and store it in count_dig[] list
# Total stores the sum of counted digits
for i in range(0,len(fibo_series)):
    for j in range(0,9):
        if leading_dig_value[i]==leadind_dig[j] :
            count_dig[j]+=1
            Total+=1

# Here we calculate the actual percentage using count_dig[] and Total
#Bendfords_predicted[] will stores the Benfords predicted percentage
#it also calculate variation between Actual & Benfords and store in variation_per_dig[]
for i in range(0,9):
    Actual_percentage.append(round((count_dig[i]/Total*100),2))
    n=log10(1+(1/leadind_dig[i]))*100
    Benfords_predicted.append(round(n,2))
    variation_per_dig.append(round(Actual_percentage[i]-Benfords_predicted[i],2))


#Following print statement print the table of above calculated data using library called tabulate 
print(tabulate({"Fibonacci":fibo_series,"Leading_Digit_value":leading_dig_value,"Leading_Digit":leadind_dig,"Count":count_dig,"Actual %":Actual_percentage,"Benfords_predicted %":Benfords_predicted,"Variation":variation_per_dig},headers="keys",tablefmt="psql",floatfmt=".2f"))


#Find the chi-test of data{actual_percentage and Benfords Predicted}
print("\nChiSquare  Comparisan as % =")
print(scipy.stats.chisquare(Actual_percentage,Benfords_predicted))

#Plot the Graph of Actual_percentage and Benfords_predicted Comparison 
N=9
ind=np.arange(N)
width=0.35
plt.bar(ind,Actual_percentage,width,label='Actual_percentage %')
plt.bar(ind+width,Benfords_predicted,width,label='Benfords_predicted %')
plt.ylabel('Percentage')
plt.title('Comparison of Actual and Benfords Percentage Prediction')
plt.xticks(ind+width/2,('1','2','3','4','5','6','7','8','9'))
plt.legend(loc='best')
plt.show()
```
